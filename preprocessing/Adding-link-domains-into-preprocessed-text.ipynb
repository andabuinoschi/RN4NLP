# Add link domains into preprocessed_text column

```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
train_data = pd.read_csv('../dataset/train.tsv', sep='\t')
validation_data = pd.read_csv('../dataset/valid.tsv', sep='\t')
test_data = pd.read_csv('../dataset/test.tsv', sep='\t')

train_augmented_data = pd.read_csv('../dataset/train_augmented.tsv', sep='\t')
```

```python
train_data[466:470]
```

```python
def add_tweet_links_to_text(dataframe):
    dataframe['preprocessed_tweet_text'] = dataframe[['preprocessed_tweet_text', 'tweet_link_domain', 'tweet_link_path']].apply(lambda row: 
                                            ' '.join([row['preprocessed_tweet_text'], row['tweet_link_domain'], row['tweet_link_path']]) if pd.notnull(row['tweet_link_domain'])
                                                                                                                                and pd.notnull(row['tweet_link_path'])
                                                                                                                                else row['preprocessed_tweet_text'], axis=1)
    return dataframe
```

```python
train_data = add_tweet_links_to_text(train_data)
validation_data = add_tweet_links_to_text(validation_data)
test_data = add_tweet_links_to_text(test_data)

train_augmented_data = add_tweet_links_to_text(train_augmented_data)
```

```python
train_data[466:470]
```

```python
train_data.to_csv('../dataset/train.tsv', sep='\t', index=False)
validation_data.to_csv('../dataset/valid.tsv', sep='\t', index=False)
test_data.to_csv('../dataset/test.tsv', sep='\t', index=False)

train_augmented_data.to_csv('../dataset/train_augmented.tsv', sep='\t', index=False)
```
